```python
columns_to_encode = ['End_time','Start_time','Location', 'Type', 'Place', 'duration_seconds']  # Replace with the names of the columns to one-hot encode

# Loop through the columns and perform one-hot encoding
for column_name in columns_to_encode:
    one_hot = pd.get_dummies(sensor_d[column_name])  # Perform one-hot encoding on the current column
    sensor_df = pd.concat([sensor_d, one_hot], axis=1)  # Concatenate the one-hot encoded columns to the original DataFrame
    sensor_datadframe = sensor_d.drop(column_name, axis=1, inplace=True)  # Drop the original column that was one-hot encoded
sensor_df
```

```python
pca = PCA()
pca.fit(df)

# Get the cumulative explained variance ratio
cumulative_var_ratio = np.cumsum(pca.explained_variance_ratio_)

# Plot the elbow curve

# Plot the elbow curve
plt.plot(cumulative_var_ratio)
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance ratio')
plt.show()
```

```python
#normalize the dataset
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
df.columns = df.columns.astype(str)
Data = scaler.fit_transform(df)
len(Data)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=3)
data_pca = pca.fit_transform(Data)
PCA_df = pd.DataFrame(data = data_pca)
PCA_df.head()
```

```python
scores = []
for k in range(2,15):
    km = KMeans(n_clusters=k,random_state=123)
    km = km.fit(PCA_df)
    scores.append(km.inertia_)
dfk = pd.DataFrame({'Cluster':range(2,15), 'Score':scores})
plt.figure(figsize=(8,5))
plt.plot(dfk['Cluster'], dfk['Score'], marker='o')
plt.xlabel('Number of clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
for i in range(2,15):
    kmeans_labels=KMeans(n_clusters=i,random_state=123).fit_predict(PCA_df)
    print("Silhouette score for {} clusters k-means : {} ".format(i,silhouette_score(Data,kmeans_labels, metric='euclidean').round(3)))
```

```python
def find_activity (row, data2):
    for _, activity_row in data2.iterrows( ) :
        if activity_row['Start_time'] <= row['Start_time'] and activity_row['End_time'] >= row['End_time']:
            return activity_row[ 'Activity']
    return None
```

```python
sensor_data['Activity'] = sensor_data.apply(lambda row: find_activity(row, label_data), axis=1)
```

```python
sensor_data
```

```python
sensor_data.dropna(subset=['Activity'],inplace=True)
```

```python
from sklearn.metrics import confusion_matrix
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
true_labels = le.fit_transform(sensor_data['Activity'])
predicted_labels = sensor_data["Clusters"]

# Create the confusion matrix
confusion_matrix = pd.crosstab(true_labels, predicted_labels, rownames=['True'], colnames=['Predicted'])
print(confusion_matrix)
```

```python
print("Activity Labels:")
for i, activity in enumerate(le.classes_):
    print(f"{activity} -> {i}")
```

```python
from sklearn.metrics import ConfusionMatrixDisplay

contingency_table = pd.crosstab(sensor_data['Activity'], predicted_labels)
confusion_matrix = contingency_table.to_numpy()

activity_labels = sorted(sensor_data['Activity'].unique())
cluster_labels = sorted(sensor_data['Clusters'].unique())

# Plot the confusion matrix using ConfusionMatrixDisplay
disp = ConfusionMatrixDisplay(confusion_matrix=confusion_matrix, display_labels=activity_labels)
disp.plot(cmap='Greens', xticks_rotation='vertical', values_format='d')

# Set the x-tick labels to match the cluster labels
plt.xticks(np.arange(len(cluster_labels)), cluster_labels)

plt.title("Confusion Matrix")
plt.show()
```

```python
accuracy = confusion_matrix.max(axis=0).sum()/len(sensor_data)
print(accuracy)
```
